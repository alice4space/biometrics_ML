```python
file = "./spiderdata/VP02/BitalinoBR.txt"
```

```python
text_strings = [] # empty array to store the last column text
with open(file) as ff:
    ss = ff.readlines() # read all strings in a string array 

for s in ss:
    text_strings.append(s.split('\t')[0]) # last column to the text array
len(text_strings)
```
